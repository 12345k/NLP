```python
import urllib.request as urllib2
from bs4 import BeautifulSoup
```

```python
def getalldoxydonkeypost(url,links):
    request = urllib2.Request(url)
    response= urllib2.urlopen(request)
    soup = BeautifulSoup(response,"lxml")

    for a in soup.findAll('a'):
#         print(a)
        try:
            url=a['href']
            title = a['title']
            
            if title == "Older Posts":
                print(title,url)
                links.append(url)
                getalldoxydonkeypost(url,links)
        except:
            title=""
    return
blogurl ="http://doxydonkey.blogspot.com"
links=[]
getalldoxydonkeypost(blogurl,links)
```

```python
def getdoxydoneytext(testurl):
    request = urllib2.Request(testurl)
    response= urllib2.urlopen(request).read().decode('utf8','ignore')
    soup = BeautifulSoup(response,"lxml")
    
    mydivs = soup.findAll("div",{"class":"post-body"})
    posts=[]
    
    for div in mydivs:
        
        text =bytes(div.find("li").text, 'utf-8')
        text=text.decode('ascii',errors='replace').replace("?","")
        text=text.replace("ï¿½","")
        print(text)
        posts.append(text)
#         posts +=map(lambda p:p.text.decode('ascii',errors='replace').replace("?",""),div.findAll("li"))
    
    return posts

    
    
    
    
    
    
```
